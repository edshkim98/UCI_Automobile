```python
df.columns = cols
data = df
```

```python
#replacing ? value with nan 
data[cols] = data[cols].replace({'?':np.nan})
data
```

```python
#Drop symboling feature
data = data.drop('symboling',axis=1)
data
```

```python
#as said in the description, it deletes all the columns which normalized-losses are null values
data = data[data['normalized-losses'].notnull()]
```

```python
#as can be seen, a lot of columns are object (non-numeric). 
#in order to use these for ml model later on, i need to change this into numerical values
print(data.dtypes)
```

```python
#a function which takes in names of all columns which have data type 'object'.
#then, it substitutes with numerical values (0,1,2...) non-numeric categorical => numeric-categorical

import enum
#x = data['engine-type'].unique()
#x[0]
data_list = ['make', 'fuel-type', 'aspiration', 'num-of-doors', 'body-style', 'drive-wheels', 'engine-location', 'engine-type',
            'num-of-cylinders', 'fuel-system']
for x in data_list:
    lst = data[str(x)].unique()
    dictionary = {}
    for i, j in enumerate(lst):
        dictionary[str(j)] = i
    data[str(x)] = data[str(x)].map(dictionary)

data
```

```python
#changes the type into either float or int from object type
#data['normalized-losses'].unique()
data_list = ['normalized-losses','bore','stroke','horsepower','peak-rpm','price']
for i in data_list:
    if '.' in data[str(i)].unique()[0]:
        data[str(i)] = data[str(i)].astype('float32')
    else:
        data[str(i)] = data[str(i)].astype('int64')
data
```

```python
#pie chart to show the relation between a feature and body-style
def pie_chart(feature):
    zero = x[x['body-style'] == 0][feature].value_counts()
    one = x[x['body-style'] == 1][feature].value_counts()
    two = x[x['body-style'] == 2][feature].value_counts()
    three = x[x['body-style'] == 3][feature].value_counts()
    four = x[x['body-style'] == 4][feature].value_counts()
    feature_index = zero.index
    feature_index2 = one.index
    feature_index3 = two.index
    feature_index4 = three.index
    feature_index5 = four.index

    explode = (0,0.1)
    fig1, ax1 = plt.subplots()
    ax1.pie(zero, labels=feature_index,autopct='%1.1f%%')
    ax1.set_title(feature+'(0 is four, 1 is two)'+' sedan ratio')
    fig2, ax2 = plt.subplots()
    ax2.pie(one, labels= feature_index2,autopct='%1.1f%%')
    ax2.set_title(feature+'(0 is four, 1 is two)'+' hatchback ratio')
    fig3, ax3 = plt.subplots()
    ax3.pie(two, labels= feature_index3,autopct='%1.1f%%')
    ax3.set_title(feature+'(0 is four, 1 is two)'+' wagon ratio')
    fig4, ax4 = plt.subplots()
    ax4.pie(three, labels= feature_index4,autopct='%1.1f%%')
    ax4.set_title(feature+'(0 is four, 1 is two)'+' hardtop ratio')
    fig5, ax5 = plt.subplots()
    ax5.pie(four, labels= feature_index5,autopct='%1.1f%%')
    ax5.set_title(feature+'(0 is four, 1 is two)'+' convertible ratio')
    #plt.legend([ax1,ax2])
    plt.show()
```

```python
#'sedan', 'hatchback', 'wagon', 'hardtop', 'convertible' 
# showing the relation between num-of-doors and body-style
pie_chart('num-of-doors')
```

```python
#data-scaling 
from sklearn.preprocessing import MinMaxScaler
minMaxScaler = MinMaxScaler()

x2 = minMaxScaler.fit_transform(x2)
```

```python

#no test dataset, so i seperated train and test set.
#there's only very little sample number, i chose test_size =0.1
x_train, x_val, y_train, y_val = train_test_split(x2, y, test_size=0.1, random_state=42)
```

```python
#first, tried with linear regression model
lin_reg = LinearRegression()
lin_reg.fit(x_train,y_train)
```

```python
#lin_reg1 = LinearRegression()
kf = KFold(n_splits=10, shuffle = True, random_state=42)
lin = cross_val_score(lin_reg, x2, y, cv=kf, scoring='neg_mean_squared_error')
print(lin)
lin_mean = lin.mean()
```

```python
pred = lin_reg.predict(x_val)
```

```python
print(r2_score(pred, y_val.values))
```

```python
mse1 = mean_squared_error(y_val, pred)
print("The mean squared error (MSE) on test set: {:.4f}".format(mse1))
```

```python

df1 = pd.DataFrame({'Actual': y_val, 'Predicted': pred, 'Difference': abs(pred - y_val)})
df1 = df1.head(10)
df1.plot(kind='bar',figsize=(15,12))
plt.grid(which='major', linestyle='-', linewidth='0.5', color='black')
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
plt.show()
```

```python
from sklearn.linear_model import ElasticNet
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn import ensemble
```

```python
#since the error was quite high, i will use gradientboosting for prediction which is much more powerful compare to linear regression
reg = ensemble.GradientBoostingRegressor(n_estimators= 1100,max_depth= 2,min_samples_split= 5,learning_rate= 0.01,loss= 'ls')
reg.fit(x_train, y_train)
#reg.fit(x2,y)
#print("train accuracy: ",reg.score(x2,y))

pred2 = reg.predict(x_val)
mse2 = mean_squared_error(y_val, pred2)
print("The mean squared error (MSE) on test set: {:.4f}".format(mse2))
```

```python
print(r2_score(pred2,y_val))
```

```python
kf = KFold(n_splits=8, shuffle = True,random_state=20)

gb_reg = cross_val_score(reg, x2, y, cv=kf, scoring ='neg_mean_squared_error')
print(gb_reg)
gb_reg_mean = gb_reg.mean()
print(gb_reg_mean)
```

```python
#booster= 'gbtree', n_estimators=1200,max_depth= 3,learning_rate= 0.01,objective = 'reg:linear',eval_metric='mae',n_jobs = 7,min_child_weight=0.2
#reg = ensemble.GradientBoostingRegressor(n_estimators= 1100,max_depth= 2,min_samples_split= 5,learning_rate= 0.01,loss= 'ls')
param_grid={'max_depth':[3,4,5],'min_samples_split':[4,5,6],
                 'n_estimators':[900,1000,1100],
                 'learning_rate':[0.01],
                 'loss': ['ls']
           }
gcv1=GridSearchCV(reg, param_grid=param_grid, cv=kf, scoring='neg_mean_squared_error', n_jobs=7)
```

```python
gcv1.fit(x2,y)
print('final params: ', gcv1.best_params_)   # 최적의 파라미터 값 출력
print('###############################')
print('best score: ', gcv1.best_score_)
```

```python
lst = [abs(gb_reg_mean), abs(lin_mean)]
lst1= ['Gradient Boost', 'Linear Regression']

plt.title('The error(mse) difference between Gradient Boost and Linear Regression')
plt.grid(which='major', linestyle='-', linewidth='0.2', color='black')
plt.grid(which='minor', linestyle=':', linewidth='0.2', color='black')
plt.bar(lst1,lst)


#As can be seen in this cross-validation score, Gradient Boost performed more than twice well compare to Linear Regression
#in terms of the mse
```
