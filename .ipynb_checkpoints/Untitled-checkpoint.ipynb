```python
df = pd.read_csv('imported.csv',header=None)
```

```python
cols = ['symboling', 'normalized-losses', 'make', 'fuel-type', 'aspiration', 'num-of-doors', 'body-style', 'drive-wheels', 
          'engine-location', 'wheel-base', 'length', 'width', 'height', 'curb-weight', 'engine-type', 'num-of-cylinders', 'engine-size', 'fuel-system', 'bore', 'stroke', 'compression-ratio', 'horsepower', 'peak-rpm', 'city-mpg', 'highway-mpg', 'price']
df.columns = cols
data = df
data
```

```python
#replacing ? value with nan 
data[cols] = data[cols].replace({'?':np.nan})
data
```

```python
#Drop symboling feature
data = data.drop('symboling',axis=1)
data
```

```python
print(data.dtypes)
```

```python
#a function which takes in names of all columns which have data type 'object'.
#then, it substitutes with numerical values (0,1,2...) non-numeric categorical => numeric-categorical
#same as label encodig

import enum
#x = data['engine-type'].unique()
#x[0]
data_list = ['make', 'fuel-type', 'aspiration', 'num-of-doors', 'body-style', 'drive-wheels', 'engine-location', 'engine-type',
            'num-of-cylinders', 'fuel-system']
for x in data_list:
    lst = data[str(x)].unique()
    dictionary = {}
    for i, j in enumerate(lst):
        dictionary[str(j)] = i
    data[str(x)] = data[str(x)].map(dictionary)

data
```

```python
#as said in the description, it deletes all the columns which normalized-losses are null values
data = data[data['normalized-losses'].notnull()]
```

```python
#as can be seen, a lot of columns are object (non-numeric). 
#in order to use these for ml model later on, i need to change this into numerical values
print(data.dtypes)
```

```python
x = data['normalized-losses']
x = x.astype('int64')
sns.distplot(x);
x.skew()

#The results show that the target variable is right-skewed, hence not normally distributed.
#
```

```python
#changes the type into either float or int from object type
#data['normalized-losses'].unique()
data_list = ['normalized-losses','bore','stroke','horsepower','peak-rpm','price']
for i in data_list:
    if '.' in data[str(i)].unique()[0]:
        data[str(i)] = data[str(i)].astype('float64')
    else:
        data[str(i)] = data[str(i)].astype('int64')
data
```
